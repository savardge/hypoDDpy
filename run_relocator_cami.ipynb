```python
import glob
from hypoddpy import HypoDDRelocator


# Init the relocator with the working directory and some necessary
# configuration values.
relocator = HypoDDRelocator(
    working_dir="/home/genevieve.savard/hypoDDpy",
    cc_time_before=0.05,
    cc_time_after=0.2,
    cc_maxlag=0.1,
    cc_filter_min_freq=1,
    cc_filter_max_freq=20,
    cc_p_phase_weighting={"Z": 1.0},
    cc_s_phase_weighting={"Z": 1.0, "E": 1.0, "N": 1.0},
    cc_min_allowed_cross_corr_coeff=0.4,
)

# Add the necessary files. Call a function multiple times if necessary.
relocator.add_event_files(
    "/home/genevieve.savard/hypoDDpy/templates/events/*.xml"
    #"/home/genevieve.savard/hypoDDpy/families_events/catalog_family_2020_03_19t05_17_44_20200310_ndet25.xml"
)
relocator.add_station_files(
    glob.glob("/home/genevieve.savard/hypoDDpy/station_files/*.xml")
)
relocator.add_waveform_files(
    glob.glob("/home/genevieve.savard/hypoDDpy/templates/waveforms/*.mseed")
    #glob.glob("/home/gilbert_lab/cami_frs/eqcorrscan/families_waveforms/*/*.mseed")
)

# Setup the velocity model. This is just a constant velocity model.
layer_tops = [(0, 1, 2.188183807),
(0.0135, 1.84, 2.925278219), 
(0.12, 2.64, 2.100238663), 
(0.227, 2.7, 2.25), 
(0.292, 2.74, 2.341880342), 
(0.359, 3, 1.986754967), 
(0.445, 3.17, 1.875739645), 
(0.5, 3.17, 1.875739645)]
relocator.setup_velocity_model(
    model_type="layered_variable_vp_vs_ratio",
    layer_tops=layer_tops)

# Start the relocation with the desired output file.
relocator.start_relocation(output_event_file="relocated_events.xml")
```
